```python
# Import necessary libraries and pytrends to get Google's daily searches

import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
trends = TrendReq()
trends
```

```python
# Find top 10 countries searched for 'data analysis'

trends.build_payload(kw_list=["data analysis"])
data = trends.interest_by_region()
data = data.sort_values(by="data analysis", ascending=False)
data = data.head(10)
data
```

```python
# Visualize the top 10 data with a bar chart

data.reset_index().plot(x="geoName",
                        y="data analysis", kind="bar")
plt.show()
```

```python
# 'data analysis' search trends

data = TrendReq(hl='en-US')
data.build_payload(kw_list=['data analysis'])
data = data.interest_over_time()
data['data analysis'].plot()
plt.title('Total Google Searches for data analysis')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
